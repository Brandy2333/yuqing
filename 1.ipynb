# 使用bs4获取数据
爬取新闻标题，在网站上对着标题右键检查，发现为“h3”，使用soup获取h3

```python
from bs4 import BeautifulSoup as bs
import requests as re
url='http://www.cuc.edu.cn/news/1901/list.htm'
try:
    r=re.get(url)
    rt=r.content
    rh=str(rt,"utf-8")
    soup=bs(rh,"html.parser")
    href=soup.find_all("h3")
    print(href)
except:
    print("error")
```

#  定义存入mysql函数
本文先在MySQL workbench上构建数据表，并设置属性，其中，title一栏应为utf-8格式，url一栏的长度应足够长

```python
from bs4 import BeautifulSoup as bs
import pymysql
def savenews(title,url):
    #pymysql.connect(数据库url，用户名，密码，数据库名)
#     db=pymysql.connect("localhost","root","cuc_news","123456",charset='utf8')
    db = pymysql.connect(host="127.0.0.1", user="root", passwd="123456", db="cucnews", port=3306,charset="utf8")
    cursor=db.cursor()
    try:
#         cursor.execute("INSERT INTO news（title,url) VALUES(%s,%s)",(title,url))
        cursor.execute('INSERT INTO news(title,url) VALUES(%s,%s)', (title,url))
        db.commit()
    except Exception as e:
        print('错误类型是',e.__class__.__name__)
        print('错误明细是',e)
        db.rollback()
    db.close()
```

# 获取信息
登录中传要闻，发现每一页的url为在list后增加数字，按此规律使用for循环进行爬取


```python
from bs4 import BeautifulSoup as bs
import requests as re
url0='http://www.cuc.edu.cn/news/1901/list'
# try:
for i in range(1,582):
    url=url0+str(i)+".htm"
    print(url)
    r=re.get(url)
    rt=r.content
    rh=str(rt,"utf-8")
    soup=bs(rh,"html.parser")
    title=soup.find_all("h3",attrs={'class','tit'})
    for t in title:
        newsurl=t.find_all('a')
        urllen=str(newsurl[0]).find("target") #find返回第一次发现字符串的下标，用于定位
        tit=t.get_text()
        url='http://www.cuc.end.cn'+(str(newsurl[0])[9:urllen-2])
        print(newsurl)
        print(url)
        print(tit)
        savenews(tit,url)
# except:
#     print("error")
```

# 将数据存入csv

```python
import requests as re
from bs4 import BeautifulSoup as bs
import csv
urlbegin = "http://www.cuc.edu.cn/news/1901/list"
try:
    f = open('cuc_news.csv','w',encoding = 'utf-8',newline='')
    csv_writer = csv.writer(f)
    csv_writer.writerow(["网址","标题"])
    for i in range(1,581):
        url = urlbegin+str(i)+".htm"
#         print(url)
        r = re.get(url)
#         print(r)
        rt = r.content
        rh = str(rt,"utf-8")
#         print(rh)
        soup = bs(rh,"html.parser")
        title = soup.find_all("h3",attrs={'class','tit'})
#         print(title)
        for t in title:
            newsurl = t.find_all('a')
            urllen = str(newsurl[0]).find("target")
            tit = t.get_text()
            url = 'http://www.cuc.edu.cn' + (str(newsurl[0])[9:urllen-2])
            csv_writer.writerow([url,tit])
    f.close()
except:
    print("error")
```
